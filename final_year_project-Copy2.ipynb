```python
class ml:
    
    
    def __init__(self,file_type,data,model,independent_variable,dependent_variable,featuring_scaling):
        self.data=data
        self.file_type=file_type
        self.model=model
        self.dependent_variable=dependent_variable
        self.independent_variable=independent_variable
        self.featuring_scaling=featuring_scaling
        
    def preprocessing(self):
        import numpy as np
        import pandas as pd
        
     
        
        if self.file_type.lower()=='csv':
            ds=pd.read_csv(self.data)
            n=ds.iloc[:,self.independent_variable].values
            x=ds.iloc[:,self.independent_variable].values
            y=ds.iloc[:,self.dependent_variable].values
        
        elif self.file_type.lower()=='xlsx':
            ds=pd.read_excel(self.data)
            n=ds.iloc[:,self.independent_variable].values
            x=ds.iloc[:,self.independent_variable].values
            y=ds.iloc[:,self.dependent_variable].values
   
    
        p=ds.isnull().sum()
        c=len(self.independent_variable)
        
        
        for i in p:
            if i!=0:
                from sklearn.impute import SimpleImputer
                imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
                imputer.fit(x[:,i])
                x[:,i] = imputer.transform(x[:,i])
           
        l=[]
        for i in range(0,c):
            if type(x[0][i])==str:
                l.append(i)   
                
        from sklearn.compose import ColumnTransformer
        from sklearn.preprocessing import OneHotEncoder
        encoder=OneHotEncoder()
        ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), l)], remainder='passthrough')
        x= np.array(ct.fit_transform(x))
        if self.model.lower()=='classification':
            from sklearn.preprocessing import LabelEncoder
            le = LabelEncoder()
            y = le.fit_transform(y)
        

        from sklearn.model_selection import train_test_split
        X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 1)
            
        if self.featuring_scaling.lower()=='yes':
            k=[] 
            for i in range(0,len(X_train[0])):
                if X_train[0][i]==0.0 and (type(X_train[0][i])==float):
                    k.append(i)
                elif X_train[0][i]==1.0 and type(X_train[0][i])==float:
                    k.append(i)
            from sklearn.preprocessing import StandardScaler
            sc = StandardScaler()
            X_train[:,len(k):] = sc.fit_transform(X_train[:,len(k):])
            X_test[:,len(k):] = sc.transform(X_test[:,len(k):])
            return X_train, X_test, y_train, y_test,self.model,self.independent_variable,n
        else:
            return X_train, X_test, y_train, y_test,self.model,self.independent_variable,n
              
```

```python
class model_selection:
    import numpy as np
    import pandas as pd
    
    
    def __init__(self,r):
        X_train, X_test, y_train, y_test,model,independent_variables,X=r.preprocessing()
        self.X_train=X_train
        self.X_test=X_test
        self.y_train=y_train
        self.y_test=y_test
        self.model=model
        self.independent_variables=independent_variables
        self.X=X
        
    
    def logistic_regression(self,l=None):
        
        
        from sklearn.linear_model import LogisticRegression
        classifier = LogisticRegression(random_state = 0)
        classifier.fit(self.X_train,self.y_train)
        y_pred = classifier.predict(self.X_test)
        from sklearn.metrics import accuracy_score
        a=accuracy_score(self.y_test, y_pred)
        return a
    
    def decision_tree_entropy_regression(self,l=None):

        from sklearn.tree import DecisionTreeClassifier
        classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
        classifier.fit(self.X_train, self.y_train)
        y_pred = classifier.predict(self.X_test)
        from sklearn.metrics import confusion_matrix, accuracy_score
        b=accuracy_score(self.y_test, y_pred)
        return b
    
    def decision_tree_gini_regression(self,l=None):

        from sklearn.tree import DecisionTreeClassifier
        classifier = DecisionTreeClassifier(criterion = 'gini', random_state = 0)
        classifier.fit(self.X_train, self.y_train)
        y_pred = classifier.predict(self.X_test)
        from sklearn.metrics import confusion_matrix, accuracy_score
        c=accuracy_score(self.y_test, y_pred)
        return c
        

    def multiple_linear_regression(self,l=None):

        from sklearn.linear_model import LinearRegression
        regressor = LinearRegression()
        regressor.fit(self.X_train,self.y_train)
        y_pred = regressor.predict(self.X_test)
        from sklearn.metrics import r2_score
        d=r2_score(self.y_test, y_pred)
        return d
    
        
    def polynomial_regression_1(self,l=None):
        from sklearn.preprocessing import PolynomialFeatures
        from sklearn.linear_model import LinearRegression
        poly_reg = PolynomialFeatures(degree = 1)
        X_poly = poly_reg.fit_transform(self.X_train)
        regressor = LinearRegression()
        regressor.fit(X_poly, self.y_train)
        y_pred = regressor.predict(poly_reg.transform(self.X_test))
        from sklearn.metrics import r2_score
        e=r2_score(self.y_test, y_pred)
        return e
        
    def polynomial_regression_2(self,l=None):
        from sklearn.preprocessing import PolynomialFeatures
        from sklearn.linear_model import LinearRegression
        poly_reg = PolynomialFeatures(degree = 2)
        X_poly = poly_reg.fit_transform(self.X_train)
        regressor = LinearRegression()
        regressor.fit(X_poly, self.y_train)
        y_pred = regressor.predict(poly_reg.transform(self.X_test))
        from sklearn.metrics import r2_score
        f=r2_score(self.y_test, y_pred)
        return f
        
    def polynomial_regression_3(self,l=None):
        from sklearn.preprocessing import PolynomialFeatures
        from sklearn.linear_model import LinearRegression
        poly_reg = PolynomialFeatures(degree = 3)
        X_poly = poly_reg.fit_transform(self.X_train)
        regressor = LinearRegression()
        regressor.fit(X_poly, self.y_train)
        y_pred = regressor.predict(poly_reg.transform(self.X_test))
        from sklearn.metrics import r2_score
        g=r2_score(self.y_test, y_pred)
        return g
        
    def polynomial_regression_4(self,l=None):
        from sklearn.preprocessing import PolynomialFeatures
        from sklearn.linear_model import LinearRegression
        poly_reg = PolynomialFeatures(degree = 4)
        X_poly = poly_reg.fit_transform(self.X_train)
        regressor = LinearRegression()
        regressor.fit(X_poly, self.y_train)
        y_pred = regressor.predict(poly_reg.transform(self.X_test))
        from sklearn.metrics import r2_score
        h=r2_score(self.y_test, y_pred)
        return h
        
    def polynomial_regression_5(self,l=None):
        from sklearn.preprocessing import PolynomialFeatures
        from sklearn.linear_model import LinearRegression
        poly_reg = PolynomialFeatures(degree = 5)
        X_poly = poly_reg.fit_transform(self.X_train)
        regressor = LinearRegression()
        regressor.fit(X_poly, self.y_train)
        y_pred = regressor.predict(poly_reg.transform(self.X_test))
        from sklearn.metrics import r2_score
        i=r2_score(self.y_test, y_pred)
        return i   
        
        
    
    def run_all(self):
        if self.model=='classification':
            self.logistic_regression()
            self.decision_tree_entropy()
            self.decision_tree_gini()
        elif self.model=='regression':
            dd=self.multiple_linear_regression()
            ee=self.polynomial_regression_1()
            ff=self.polynomial_regression_2()
            gg=self.polynomial_regression_3()
            hh=self.polynomial_regression_4()
            ii=self.polynomial_regression_5()
            return dd,ee,ff,gg,hh,ii,self.model
```
